# TPC-H DuckDB 1回実行

TPC-Hの1~22までのクエリをHTTPリクエスト経由のDuckDBで実行したときのレスポンスタイムを可視化する

```python
import requests
from time import sleep

result = []

for i in range(1, 23):
    response = requests.get(f"http://localhost:8000?q={i}")
    response_data = response.json()
    # print(i, response.status_code, response_data["elapsed"])
    result.append(response_data["elapsed"])

    # 1秒sleepしてサーバを休ませる
    sleep(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, len(result) + 1)

# result配列をプロット
plt.bar(x, result)
plt.xlabel('Request Index')
plt.ylabel('Elapsed Time')
plt.title('Elapsed Time for Each Request')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
